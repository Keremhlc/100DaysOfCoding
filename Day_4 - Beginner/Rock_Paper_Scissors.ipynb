```python
import random

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

#Write your code below this line ðŸ‘‡

player_cho = int(input("What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors.:"))

if player_cho == 0:
  print("Player Choice:\n", rock)
elif player_cho == 1:
  print("Player Choice:\n", paper)
else:
  print("Player Choice:\n", scissors)

machine_cho = random.randint(0, 2)

if machine_cho == 0:
  print("Machine Choice:\n", rock)
elif machine_cho == 1:
  print("Machine Choice:\n", paper)
else:
  print("Machine Choice:\n", scissors)


if player_cho == 0:
  if machine_cho == 0:
    print("Play Againg")
  elif machine_cho == 1:
    print("Machine Wins")
  else:
    print("Player Wins")

elif player_cho == 1:
  if machine_cho == 0:
    print("Player Wins")
  elif machine_cho == 1:
    print("Play Again")
  else:
    print("Machine Wins")
    
else:
  if machine_cho == 0:
    print("Machine Wins")
  elif machine_cho == 1:
    print("Player Wins")
  else:
    print("Play Again")
```
